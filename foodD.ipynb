```python
from ultralytics import YOLO
 
# Load the model.
model = YOLO('yolov8n.pt')
 
results = model.train(
    data='data.yaml',
    imgsz=1024, # Increase image size
    epochs=50, # Increase number of epochs
    batch=2, # Decrease batch size
    name='yolov8n_custom',
    augment=[   
        {'name': 'RandomCrop', 'width': [0.5, 0.8], 'height': [0.5, 0.8], 'p': 0.5},
        {'name': 'RandomFlip', 'horizontal': True, 'vertical': False, 'p': 0.5},
        {'name': 'ColorJitter', 'brightness': [0.8, 1.2], 'contrast': [0.8, 1.2], 'saturation': [0.8, 1.2], 'hue': 0.1},
        {'name': 'GaussianBlur', 'kernel_size': (3, 3), 'sigma': (0.1, 2.0)},
        {'name': 'Rotate', 'angle': [-10, 10], 'p': 0.5},
        {'name': 'RandomScale', 'scale_ratio': [0.8, 1.2], 'p': 0.5},
        {'name': 'CutOut', 'n_holes': [1, 3], 'length': [0.1, 0.3]},
        {'name': 'Mosaic', 'p': 0.5, 'min_ratio': 0.5, 'max_ratio': 2},
        {'name': 'Blur', 'blur_limit': [3, 7]},
        {'name': 'GaussNoise', 'var_limit': (5.0, 30.0)}],
    optimizer='Adam',
    save_period=2
)
```
